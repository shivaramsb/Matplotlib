```python
import matplotlib.pyplot as plt

x = list(range(10))
y = [i**2 for i in x]
```

**1.Plot x vs. y using a line plot.**

```python
plt.plot(x, y, label='Line Plot')
```

**2. Add a title to the plot.**

```python
plt.title('Basic Plotting')
```

**3. Label the x-axis as "X-axis".**

```python
plt.xlabel('X-axis')
```

**4. Label the y-axis as "Y-axis".**

```python
plt.ylabel('Y-axis')
```

**5. Change the color of the line to red.**

```python
plt.plot(x, y, color='red', label='Red Line Plot')
```

**6. Add markers to the line plot.**

```python
plt.plot(x, y, marker='o', label='Line Plot with Markers')
```

**7. Save the plot as a PNG file.**

```python
plt.savefig('basic_plot.png')
```

**8. Display the grid on the plot.**

```python
plt.grid(True)
```

**9. Plot x vs. y using a scatter plot on the same figure.**

```python
plt.scatter(x, y, label='Scatter Plot')
```

**10. Add a legend to differentiate between the line and scatter plots**

```python
plt.legend()
```
