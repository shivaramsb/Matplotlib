```python
import matplotlib.pyplot as plt
import numpy as np

# Set seed for reproducibility
np.random.seed(0)

# Generate random student scores between 50 and 100
scores = np.random.randint(50, 101, size=100)
```

**1.Create a histogram of the scores.**

```python
plt.hist(scores, bins=10, color='skyblue', edgecolor='black')
```

**2. Add a title to the plot.**

```python
plt.title('Student Scores Distribution')
```

**3. Label the x and y axes**

```python
plt.xlabel('Scores')
plt.ylabel('Frequency')
```

**4.Change the color of the bins (already done in 1).**

**5. Set the number of bins to 10 (already done in 1).
**

**6. Add a grid to the plot.**

```python
plt.grid(True)

```

**7. Add a line for the mean score.**

```python
mean_score = np.mean(scores)
plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=1, label='Mean')
```

**8.Add a legend for the mean line.**

```python
plt.legend()
```
