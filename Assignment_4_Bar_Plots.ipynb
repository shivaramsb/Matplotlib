```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [150, 200, 300, 250]
```

**1.Create a bar plot for sales data.**

```python
plt.bar(products, sales, color='purple')
```

**2. Add a title to the plot.**

```python
plt.title('Sales Data')
```

**3. Label the x and y axes.**

```python
plt.xlabel('Products')
plt.ylabel('Sales')
```

**4.Change the color of the bars (already done in 1).**

**5. Add value labels on top of each bar.**

```python
for i, v in enumerate(sales):
    plt.text(i, v + 10, str(v), ha='center')
```

**6. Rotate the x-axis labels by 45 degrees.**

```python
plt.xticks(rotation=45)

```

**7. Change the width of the bars.**

```python
plt.bar(products, sales, width=0.5, color='purple')
```

**8.Add a grid to the plot.**

```python
plt.grid(True)
```

**9. Save the plot as a PDF file.**

```python
plt.savefig('bar_plot.pdf')
```

**10. Display the plot.**

```python
plt.show()
```
