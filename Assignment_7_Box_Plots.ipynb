```python
import matplotlib.pyplot as plt

math_scores = [78, 85, 88, 92, 76, 84, 89, 93, 91, 77]
science_scores = [80, 83, 88, 85, 79, 90, 86, 92, 87, 81]
english_scores = [74, 82, 81, 90, 78, 85, 89, 91, 88, 77]

data = [math_scores, science_scores, english_scores]
```

**1.Create a box plot for the scores in different subjects.**

```python
plt.boxplot(data, notch=True, patch_artist=True, showmeans=True, meanline=True,
            boxprops=dict(facecolor='lightblue'))
```

**2. Add a title to the plot.**

```python
plt.title('Scores Distribution by Subject')
```

**3. Label the x and y axes.**

```python
plt.xlabel('Subjects')
plt.ylabel('Scores')
```

**4. Add grid lines to the plot.**

```python
plt.grid(True)
```

**5. Save the plot as a PDF file.**

```python
plt.savefig('box_plot.pdf')
```

**6. Display the plot.**

```python
plt.show()
```

**7. Add individual data points as a scatter plot on top of the box plots.**

```python
plt.boxplot(data, notch=True, patch_artist=True, showmeans=True, meanline=True,
            boxprops=dict(facecolor='lightblue'))
for i in range(len(data)):
    y = data[i]
    x = [i + 1] * len(y)
    plt.scatter(x, y, color='black')
plt.title('Scores Distribution by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(True)
plt.savefig('box_plot_with_scatter.pdf')
plt.show()
```
