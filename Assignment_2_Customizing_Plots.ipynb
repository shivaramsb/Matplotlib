```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperatures = [0.5, 1.0, 4.5, 9.0, 14.5, 17.5, 20.0, 19.0, 15.0, 10.0, 4.5, 1.0]
```

**1.Plot months vs. temperatures using a line plot.**

```python
plt.plot(months, temperatures, label='Temperature')
```

**2. Set the figure size to 10x5 inches.**

```python
plt.figure(figsize=(10, 5))
```

**3. Use a different color and linestyle for the line.**

```python
plt.plot(months, temperatures, color='green', linestyle='--', label='Green Dashed Line')
```

**4. Add markers to the line plot.**

```python
plt.plot(months, temperatures, marker='o', label='Line Plot with Markers')
```

**5. Add a title and axis labels.**

```python
plt.title('Monthly Temperatures')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
```

**6. Rotate the x-axis labels by 45 degrees.**

```python
plt.xticks(rotation=45)
```

**7. Change the y-axis range to [-5, 25].**

```python
plt.ylim([-5, 25])
```

**8. Add a grid to the plot.**

```python
plt.grid(True)
```

**9. Annotate the highest temperature point.**

```python
plt.annotate('Highest Temp', xy=('Jul', 20.0), xytext=('Jun', 22.0),
             arrowprops=dict(facecolor='black', shrink=0.05))
```

**10. Save the plot as a PDF file.**

```python
plt.savefig('custom_plot.pdf')
```
