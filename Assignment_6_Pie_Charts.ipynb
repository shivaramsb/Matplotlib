```python
import matplotlib.pyplot as plt

companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [35, 25, 20, 20]
```

**1.Create a pie chart of market share data.**

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
```

**2. Add a title to the plot.**

```python
plt.title('Market Share of Companies')
```

**3. Label the x and y axes**

```python
plt.xlabel('Scores')
plt.ylabel('Frequency')
```

**4.Add a legend outside the pie chart.**

```python
plt.pie(market_share, explode=explode, labels=companies, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.legend(companies, loc="best")
```

**5. Ensure the pie chart is a circle (achieved by aspect ratio).**

```python
plt.axis('equal')
```
