```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

**1.Create a 3D scatter plot of the points.**

```python
sc = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')
```


 **2. Add a title to the plot.**

```python
ax.set_title('3D Scatter Plot')
```

**3. Label the x, y, and z axes.**

```python
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_ylabel('Z-axis')
```

**4.  Add a color bar based on the z-values.**

```python
plt.colorbar(sc)
```

**5.Rotate the plot to a different viewing angle.**

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')
ax.set_title('3D Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_ylabel('Z-axis')
plt.colorbar(sc)
ax.view_init(elev=30, azim=120)
```

**6. Add grid lines to the plot.**

```python
ax.grid(True)

plt.savefig('3d_scatter_rotated.png')
plt.show()
```
